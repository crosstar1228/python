## 3. os.rename으로 수정 완료


```python
path = '/content/drive/MyDrive/Colab Notebooks/[해커톤]DeepASMR/Dataset/bird'
path2 = path + '/bird_2'
path3 = path + '/bird_3'
path4 = path + '/bird_4'
```

```python
import os 
file_names = os.listdir(path2)
file_names
```

```python
i=724
for name in file_names:
  src = os.path.join(path2, name)
  dst = str(i) + '.wav'
  dst = os.path.join(path2, dst)
  os.rename(src,dst)
  i+=1
```

```python
i=934
for name in file_names:
  src = os.path.join(path3, name)
  dst = str(i) + '.wav'
  dst = os.path.join(path3, dst)
  if name.endswith('.wav'):
    os.rename(src,dst)
  i+=1
```
