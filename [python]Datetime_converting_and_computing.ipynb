# Dateutil 패키지
 - `strptime` 은 문자열에 맞는 형식을 사용자가 제공해야 되는데, 이건 `parse` 함수 쓰면 자동으로 찾아서 `datetime` 객체를 반환해 줌

```python
from dateutil.parser import parse
```

```python
parse('2016-04-16')
```

```python
parse("APR 16, 2016 04:05:32 PM")
```

- 월/일이 둘다 12보다 작을경우, 먼저 나온 수를 월로 판단함

```python
parse('6/7/2016')
```

## 날짜, 시간 연산
- `datetime`클래스 간 차이 연산
  -> `timedelta` 클래스 객체로 반환

```python
dt1 = dt.datetime(2016, 2, 19, 14)
dt2 = dt.datetime(2016, 1, 2, 13)
td = dt1 - dt2
td
```

### `timedelta` 클래스의 attribute and method
  - attribute : `days`, `seconds`, `microseconds`
  - method : `total_seconds`(모든 attribute을 초단위로 모아 반환)

```python
td.days, td.seconds, td.microseconds
```

```python
td.total_seconds()
```

`datetime` 과` timedelta `덧셈도 가능

```python
t0 = dt.datetime(2018, 9, 1, 13)
d = dt.timedelta(days=90, seconds=3600)
t0 + d
```

datetime 연산은 초단위 연산이므로, 필요에 따라 월단위 연산이 가능한 `dateutil` 패키지의 `relativedelta` 모듈을 import해서 사용

```python
from dateutil.relativedelta import relativedelta

t0 + relativedelta(months=2)
```
